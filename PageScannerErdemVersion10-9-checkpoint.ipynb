```python
import cv2 
from google.colab.patches import cv_imshow
import numpy as np
```

Taking image part

```python
image = cv2.imread('/content/IMG_0857.jpg')
```

```python
image = cv2.resize(image, (1080, 1980))
orig = image.copy()
```

Making image smoother

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#image = cv2.cv2tColor(image, cv2.COLOR_BGR2HSV )

image = cv2.GaussianBlur(image, (3,3), 0)
#image = cv2.medianBlur(image,5)
```

Doing Edge detection with Canny

```python
edged = cv2.Canny(image, 15, 50,apertureSize=5,L2gradient =True)
#orig_edged = edged.copy()
```

```python
(contours, _) = cv2.findContours(edged, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
contours = sorted(contours, key=cv2.contourArea, reverse=True)
```

```python
for c in contours:
    p = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.02 * p, True)

    if len(approx) == 4:
        target = approx
        break
```

```python
def rectify(h):
    h = h.reshape((4,2))
    hnew = np.zeros((4,2),dtype = np.float32)

    add = h.sum(1)
    hnew[0] = h[np.argmin(add)]
    hnew[2] = h[np.argmax(add)]

    diff = np.diff(h,axis = 1)
    hnew[1] = h[np.argmin(diff)]
    hnew[3] = h[np.argmax(diff)]

    return hnew
```

Resize our captured and cutted paper area

```python
approx = rectify(target)
pts2 = np.float32([[0,0],[1080,0],[1080,1980],[0,1980]])
```

```python
M = cv2.getPerspectiveTransform(approx,pts2)
dst = cv2.warpPerspective(orig,M,(1080,1980))

cv2.drawContours(image, [target], -1, (0, 255, 0), 2)
dst = cv2.cvtColor(dst, cv2.COLOR_BGR2GRAY)
```

In this part we do Treshholds for making image look more scanned 

```python

thresh2 = cv2.adaptiveThreshold(dst,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,8)
thresh3 = cv2.adaptiveThreshold(thresh2,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,3)
ret2,thresh4 = cv2.threshold(thresh3,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret,thresh1 = cv2.threshold(thresh4,0,255,cv2.THRESH_BINARY)
```

Get rid of dots

```python
cleanNoise = cv2.fastNlMeansDenoising(thresh1,None,55,10,7)
st=cv2.getStructuringElement(cv2.MORPH_RECT , (1,1))
closed=cv2.morphologyEx(thresh1, cv2.MORPH_CLOSE, st)
out_gray=cv2.divide(thresh1, closed, scale=255)
out_binary=cv2.threshold(out_gray, 0, 255, cv2.THRESH_OTSU )[1] 
```
